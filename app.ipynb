```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def apply_fft(signal, fs, t):
    N=len(t)
    fft_result = np.fft.fft(signal)
    freqs = np.fft.fftfreq(N, 1/fs)
    return freqs, fft_result    
```

```python
def apply_ifft(signal, t):
    N=len(t)
    ifft_result = np.fft.ifft(signal)
    return ifft_result
```

```python
def draw_signal(x1,x2,x3,t):
    plt.figure(figsize=(12, 3))
    plt.plot(t, x1, label='x1(t) = cos(2πf1t)')
    plt.plot(t, x2, label='x2(t) = 3cos(2πf2t)')
    plt.plot(t, x3, label='x3(t) = x1(t) + x2(t)', linestyle='dashed')  
    plt.title('Zaman Domeni Sinyalleri')
    plt.xlabel('Zaman [s]')
    plt.ylabel('Genlik')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def draw_fft(fft, freqs, fs):
    plt.figure(figsize=(10, 3))
    plt.stem(freqs, fft)
    plt.title("Frekans Domeni (FFT)")
    plt.xlabel("Frekans (Hz)")
    plt.ylabel("Genlik")
    plt.xticks(np.arange(-fs/2, fs/2, 2))
    plt.grid()

    plt.tight_layout()
    plt.show()
```

```python
def draw_ifft(ifft, t):
    plt.figure(figsize=(10, 3))
    plt.plot(t, ifft, label='x3(t) = x1(t) + x2(t)', linestyle='dashed')
    plt.title("Zaman Domeni (IFFT)")
    plt.xlabel("Zaman (s)")
    plt.ylabel("Genlik")
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
fs = 30
t = np.arange(0, 1, 1/fs)
f1 = 5
f2 = 12

x1_signal = np.cos(2*np.pi*f1*t)
x2_signal = 3*np.cos(2*np.pi*f2*t)
x3_signal = x1_signal + x2_signal
draw_signal(x1_signal, x2_signal, x3_signal, t)
freqs, fft = apply_fft(x3_signal, fs, t)
draw_fft(fft, freqs, fs)
ifft = apply_ifft(fft, fs, t)
draw_ifft(ifft, t)
```
